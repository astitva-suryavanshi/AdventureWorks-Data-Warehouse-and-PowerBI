```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("02_AdventureWorks Customer Lookup.csv",encoding="latin1")
print(df)
```

```python
df = df[df["CustomerKey"].astype(str).str.isnumeric()]
```

```python
df.tail(10)
```

```python
print(df["CustomerKey"].dtype)
```

```python
df["CustomerKey"] = pd.to_numeric(
    df["CustomerKey"],
    errors="coerce"
)
```

```python
df = df.dropna(subset=["CustomerKey"])
```

```python
df["CustomerKey"] = df["CustomerKey"].astype(int)
```

```python
print(df)
```

```python
df["FirstName"] = df["FirstName"].str.title()

df["LastName"] = df["LastName"].str.title()
```

```python
print(df["Prefix"].unique())
```

```python
df["Prefix"] = (
    df["Prefix"]
    .str.upper()
    .str.strip()
)
```

```python
df["BirthDate"] = pd.to_datetime(
    df["BirthDate"],
    errors="coerce"
)
```

```python
print(df["BirthDate"].dtype)
```
