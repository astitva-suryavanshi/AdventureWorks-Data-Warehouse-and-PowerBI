```python
import pandas as pd
import numpy as np
```

```python
sales2020 = pd.read_csv("06_AdventureWorks Sales Data 2020.csv", encoding="latin1")
sales2020
```

```python
sales2021 = pd.read_csv("07_AdventureWorks Sales Data 2021.csv", encoding="latin1")
sales2021
```

```python
sales2022 = pd.read_csv("08_AdventureWorks Sales Data 2022.csv", encoding="latin1")
sales2022
```

```python
sales = pd.concat(
    [sales2020, sales2021, sales2022], ignore_index= True)
```

```python
sales
```

```python
sales["OrderDate"] = pd.to_datetime(
    sales["OrderDate"]
)

sales["DateKey"] = sales["OrderDate"].dt.strftime("%Y%m%d")
```

```python
sales
```

```python
sales.columns
```

```python
sales.to_excel("09_AdventureWorks Sales Data 2020-2022.xlsx", index=False)
```
